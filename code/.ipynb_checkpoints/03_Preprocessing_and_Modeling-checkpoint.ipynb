#### Defined functions:


```python
# function for printing value counts
def val_count(df, columns):
    """ for value counts"""
    
    print (df[columns].value_counts(dropna=False))
```

```python
# fucntion for viewing nulls
def viewnull (df, columns):
    """for viewing 'NaN' entries by column and dataframe"""
    
    return df[df[columns].isnull()]
```

```python
# function for viewing unique
def viewuni(df, columns):
    """for viewing unique entries by column and dataframe"""
    
    print(df[columns].unique())
```

```python
# function for replacing values
def replace_val(df, columns, old_value, new_value):
    """ for replacing values in columns"""
    
    df[columns] = df[columns].replace({old_value: new_value})
```

```python
# mask function
def and_test(df, column_1, value_1, column_2, value_2 ):
    """ returns row where column_1 is value_1 and column_2 is not value 2"""
    
    mask= (df[column_1] == value_1) & (df[column_2] !=value_2)
    return df.loc[mask]
```

```python
# function for replacing specific values
def replace_sval(df, index, columns, value):
    """for replacing specific values by index and column"""
    
    df.at[index, columns] = value
    return print(df[columns][index])
```

```python
# function for null and not null pair testing
def null_test(df, column_1,column_2):
    """ returns row where column_1 is null, column_2 is not null"""
    
    mask = (df[column_1].isnull()) & (df[column_2].notnull())
    return df.loc[mask]
```

```python
# boxplot subplots function
def plotbox(df, column_x, column_y, subplot_no, title):
    """for plotting boxplots in subplots"""
    
    # indicate subplot
    ax1 = plt.subplot(subplot_no) 
    
    # set title
    ax1.set_title(title) 
    
    # boxplot
    sns.boxplot(data=df,x=column_x, y=column_y, color= 'salmon')
    
    # mean line
    plt.axhline(np.median(df[column_y]), color='black',linestyle='--')
    
    # indicate lengend, top right
    plt.legend([column_y +" Mean"], loc="upper right");
```

```python
# distplot function
def plotdist (df, columns, subplot_no, title):
    """for plotting distplot in subplots"""
    
    # indicate subplot
    ax1 = plt.subplot(subplot_no) 
    
    # set title
    ax1.set_title(title) 
    
    # distribution plot, with kde
    sns.distplot(df[columns], kde=True); 
```

```python
# barplot function
def plotbar(df, column_x, column_y, subplot_no, title):
    """for plotting barplots in subplots"""
    
    # indicate subplot
    ax1 = plt.subplot(subplot_no) 
    
    # set title
    ax1.set_title(title)
    
    # barplot
    sns.barplot(x=column_x, y=column_y, data=df, color ='tomato', estimator=np.median); 
```

```python
# function to creat heatmaps with mask
def mapheat(df, size):
    """ function to create custom heatmap"""
    
    # establish size of figure
    plt.figure(figsize=(size,size))

    # find corr
    corr = df.corr()
    
    # create mask for upper triangle
    mask1 = np.zeros_like(corr)
    mask1[np.triu_indices_from(mask1)] = True

    # plot the heatmap with mask
    with sns.axes_style("white"):
        sns.heatmap(corr, mask=mask1, vmax=0.3, square=True, annot=True, cmap='vlag');
```

```python
# barplot function
def plotreg(df, column_x, column_y, subplot_no, title):
    """for plotting regplots in subplots"""
    
    # indicate subplot
    ax1 = plt.subplot(subplot_no)
    
    # set title
    ax1.set_title(title)
    
    # regplot
    sns.regplot(x=column_x, y=column_y, data=df, color='salmon'); 
```

```python
# function to map change in values in columns by dict:
def val_map(df, columns, dictionary):
    """ to change values in columns by mapping a dictionary"""
    
    df[columns]= df[columns].map(dictionary)
    print (df[columns].value_counts(dropna=False))
```

```python
# function for corr only showing saleprice corr
def corrsort(df, columns):
    """ for returning corr single column"""
    
    corr_all = df.corr().abs()
    unstacked = corr_all.unstack()
    sort = unstacked.sort_values(ascending=False)
    
    return sort[columns]
```

```python
# view residuals
def view_res(pipeline, X_train, X_test, title):
    """view residuals in plot"""
    
    # creating data points
    res_test = pd.DataFrame({'y value': y_test, 'residual': 
                             y_test - pipeline.predict(X_test),'label' : "test"})
    res_train = pd.DataFrame({'y value': y_train, 'residual': 
                              y_train - pipeline.predict(X_train),'label' : "train"})
    residuals = pd.concat([res_test, res_train])
    
    # lmplot
    lmplot = sns.lmplot(x = 'y value', y = 'residual', hue = 'label', 
                        data = residuals, lowess = True, scatter_kws={"alpha": 0.3})
    # set title
    plt.title(title);
```

```python
# function for features + train test split
def feat_train(df, y_column):
    """ creating X,y features and fitting them into train,test, split function"""
    
    features = [col for col in df._get_numeric_data().columns if col != y_column if col!= 'id']
    
    # setting X and y
    global X
    global y
    X = df[features]
    y = df[y_column]

    
    # train test split
    global X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    random_state = 88)
  
```

```python
# function for plotting barplot for coef
def coef_plot(model, X, title):
    """ creating a barplot to show coef frome pipeline"""
    
    # extract and sort coef
    model_coef= gs_enet.best_estimator_.named_steps['enet'].coef_
    coefs = pd.DataFrame({'name': X.columns, 'coef' :model_coef}).sort_values('coef')

    # plot
    sns.pointplot(y="name", x="coef",
                  data=coefs, ci=None, color = 'tomato')
    sns.barplot(y = "name", x= "coef", data=coefs, ci=None, color = 'tomato', alpha = 0.2)
    plt.title(title)

    
```

#### 4.4 Feature Engineering

We combine some of the features we are interested in and try giving them different weight.
I use the corrsort function to help me check if the correlation has gone up or down with the dropping and addition or product of each column used. 

the process:
1. create a list of columns
2. add the list into the df as a column
3. run the corrsort function to view correlation
4. drop or add columns into the list 
5. repeat steps 2-5


See: [Defined functions:](#Defined-functions:)


Back to: [3.0 Preprocessing and Feature Engineering](#3.0-Preprocessing-and-Feature-Engineering)

```python
# view data
train.head()
```

```python
# list sorted corr 
corrsort(train, 'saleprice')
```

```python
# create a Total Area feature column
# list all columns to be added together
ta_list =['lot_area', 'mas_vnr_area', 'total_bsmt_sf', 'gr_liv_area', 'garage_area', 
          'wood_deck_sf', 'open_porch_sf', 'enclosed_porch', 'porch_3ssn', 'screen_porch',
          'pool_area']
# create and add value into new column
train['total_area'] = train[ta_list].sum(axis=1)
```

```python
# create a Total Quality feature column
# list all columns to be added together
tq_list =['overall_qual', 'exter_qual', 'bsmt_qual', 'kitchen_qual', 
          'fireplace_qu', 'garage_qual', 'pool_qc', 'heating_qc']
# create and add value into new column
train['total_qual'] = train[tq_list].sum(axis=1)
```

```python
# create a Total Area and Quality feature column

# create and add value into new column
train['total_area_qual'] = train['total_area'] * train['total_qual']
```

```python
train_years = train['year_built']*train['year_remod/add']
```

```python
# create a Total Quality 2 feature column to maximise corr
# list all columns to be added together
tq2_list =['overall_qual', 'exter_qual', 'bsmt_qual', 'kitchen_qual', 
          'fireplace_qu', 'pool_qc',]
# create and add value into new column
train['total_qual_2'] = train[tq2_list].sum(axis=1)*train_years
```

```python
# create a Total Area 2 feature column to maximise corr
# list all columns to be added together
ta2_list =['mas_vnr_area', 'total_bsmt_sf', 'gr_liv_area', 'garage_area', 
          'wood_deck_sf', 'open_porch_sf', 'porch_3ssn', 'screen_porch']
# create and add value into new column
train['total_area_2'] = train[ta2_list].sum(axis=1)*train_years
```

```python
# create a Total Area and Quality 2 feature column to maximise corr

# create and add value into new column
train['total_area_qual_2'] = train['total_area_2'] * train['total_qual_2']*train_years
```

```python
# create an Garage overall feature column
# list all columns to be multiplied together
go_list =['garage_finish', 'garage_cars', 'garage_qual', 'garage_cond',
          'garage_yr_blt', 'garage_area']
# create and add value into new column
train['garage_overall'] = train[go_list].prod(axis=1)*train_years
```

```python
# create an Basement overall feature column
# list all columns to be added together
bo_list =['bsmt_qual', 'bsmt_cond', 'bsmt_exposure', 'bsmtfin_type_1', 
          'bsmtfin_sf_1' ,'bsmtfin_sf_2' ,'bsmt_unf_sf']
# create and add value into new column
train['bsmt_overall'] = (train[bo_list].sum(axis=1))*train_years
```

```python
# create an Bed + Bath feature column
# list all columns to be added together
bb_list =['full_bath', 'totrms_abvgrd', 'bedroom_abvgr']
# create and add value into new column
train['bed_bath'] = (train[bb_list].sum(axis=1))
```

```python
# create an Basement + Masonary veneer + garage column
# list all columns to be added together
bmg_list =['mas_vnr_area', 'total_bsmt_sf', 'garage_area']
# create and add value into new column
train['bsmt_mas_garage_area'] = (train[bb_list].sum(axis=1))
```

```python
# create an top neighbourhood column
# list all columns to be added together
tn_list =['neighborhood_NoRidge', 'neighborhood_StoneBr', 'neighborhood_NridgHt']
# create and add value into new column
train['top_neighbh'] = (train[tn_list].sum(axis=1))
```

```python
# check
corrsort(train, 'saleprice')
```
